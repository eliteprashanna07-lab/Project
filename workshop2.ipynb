```python
#Exercise3
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

def twin_primes(limit):
    twins = []
    for i in range(3, limit, 2):
        if is_prime(i) and is_prime(i + 2):
            twins.append((i, i + 2))
    return twins

print("Twin primes less than 1000:")
print(twin_primes(1000))
```

```python
#Exercise4
def sumOfDivisors(n):
    if n < 1:
        return 0
    total = 0
    divisor = 1
    while divisor < n:
        if n % divisor == 0:
            total += divisor
        divisor += 1
    return total
print(sumOfDivisors(7))
```

```python
# Exercise - 5
def sumOfDivisors(n): # Using the same function from exercise - 4
    total = 0

    for i in range(1, n):
        if n % i == 0:
            total += i

    return total
n = 7 # Going from 7 because we need to find the perfect number bigger than 6
while True:
  if sumOfDivisors(n) == n: # Perfect number condition
    print("The next perfect number is: ", n)
    break
  n += 1
```

```python
# Exercise - 6
marks = [20, 30, 35, 55, 70, 80, 90]

# Mean
total = 0
for m in marks:
    total += m
mean = total / len(marks)

# Median
n = len(marks)
if n % 2 == 1:
    median = marks[n // 2] # For even number of data
else:
    median = (marks[n // 2 - 1] + marks[n // 2]) / 2 # For odd number of data

print("Mean --> ", mean)
print("Median --> ", median)
```

```python
# Exercise - 7
def convert_to_binaryNum(n):
  binary = "" # Initialising an empty string for binary
  while n / 2 == 0:
    binary += str(n % 2) # Storing the remainder in binary string
    n = n // 2
  return binary[::-1] # Reversing the String
print(convert_to_binaryNum(11))
```

```python
# Exercise - 8
def prodDigits(n):
  prod = 1 # Initialising to 1 because we have to multiply
  while n > 0:
    prod *= n % 10 # Multiplying with the remainder
    n = n // 10
  return prod
print(prodDigits(27))
```

```python
# Exercise - 9
def multiply(*args): # *args gets multiple arguments
  prod = 1 # Initialising to 1 because we have to multiply
  for i in args:
    prod *= i # Multiplying every argument
  return prod
print(multiply(1,2,3,4,5))
print(multiply(12, 13, 14))
```

```python
# Example - 3
set1 = {10, 20, 30, 40, 50}
set2 = {30, 40, 50, 60, 70}
set3 = {10, 20, 30}

# a. Length of set1
print("a. Length of set1:", len(set1))

# b. Sum of elements in set2
total = 0
for i in set2:
    total += i
print("b. Sum of set2:", total)

# c. Union
print("c. Union:", set1 | set2)

# d. Intersection
print("d. Intersection:", set1 & set2)

# e. set1 - set2
print("e. set1 - set2:", set1 - set2)

# f. set2 - set1
print("f. set2 - set1:", set2 - set1)

# g. Check subset
print("g. set3 subset of set1:", set3.issubset(set1))
print("   set3 subset of set2:", set3.issubset(set2))

# h. Complement of set1 with U = {1,...,50}
U = set()
for i in range(1, 51):
    U.add(i)

complement = U - set1
print("h. Complement of set1:", complement)

# i. Check if 15 is in set1
print("i. Is 15 in set1?", 15 in set1)

# j. Add 60 to set1
set1.add(60)
print("j. set1 after adding 60:", set1)

# k. Remove 60 from set2
set2.remove(60)
print("k. set2 after removing 60:", set2)

# l. Remove all elements from set3
set3.clear()
print("l. set3 after removing all elements:", set3)

# m. Find maximum and minimum in set2
max_val = max(set2)
min_val = min(set2)

print("m. Maximum in set2:", max_val)
print("   Minimum in set2:", min_val)
```
